# ANOVA F-value

Import SelectKBest to select the number of features and f_classif to calculate the AVOVA F-value score

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
```

```python
# to get all features ranking
selector_f = SelectKBest(f_classif, k='all').fit(data,y)
```

Create a dataframe of the feature names and their ANOVA F-value scores

```python
df_scores = pd.DataFrame(selector_f.scores_)
df_columns = pd.DataFrame(data.columns)
feature_scores = pd.concat([df_columns, df_scores],axis=1)
feature_scores.columns = ['Feature_Name','Score']
```

```python
feature_scores
```

Rank the features according to their score

```python
data_use=feature_scores.nlargest(13116,'Score') 
```
