```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_fi = pd.read_csv("FI_values_cv_only_crf.csv")
```

```python
data_fi
```

```python
data_fi = data_fi.drop(['Unnamed: 0'], axis=1)
```

```python
data_fi
```

```python
data_fi['mean'] = round(data_fi.mean(axis=1),3)
```

```python
data_fi
```

```python
data_for_names = pd.read_csv("cv_x_under.csv")
```

```python
data_for_names = data_for_names.drop(['Unnamed: 0'], axis=1)
```

```python
data_for_names.columns
```

```python
data_for_names.columns[45:65]
```

```python
df_scores = pd.DataFrame(data_fi['mean'])
```

```python
df_columns = pd.DataFrame(data_for_names.columns[45:65])
```

```python
feature_scores = pd.concat([df_columns, df_scores], axis =1)
```

```python
feature_scores.columns = ['Feature_Name', 'Score']
```

```python
feature_scores
```

```python
# ordering
```

```python
data_use = feature_scores.nlargest(19, 'Score')
```

```python
data_use
```

```python
feature_scores.shape
```

```python
data_use.shape
```

```python
data_use.to_csv("feature_imp_only_crf_ordered.csv")
```

```python
data_use_15 = data_use.iloc[:15]
```

```python
data_use_15
```

```python
plotting_names = data_use_15['Feature_Name']
plotting_values = data_use_15['Score']

fig = plt.figure(figsize = (8, 4))
plt.bar(plotting_names, plotting_values, color ='c',
        width = 0.8)
plt.xlabel("Features")
plt.xticks(rotation = 90)
plt.ylabel("Importance values")
plt.title("Feature Importance for CRF (including BP) for CVD")
plt.show()
```
