```python
# undersampling
```

```python
from imblearn.under_sampling import RandomUnderSampler
from collections import Counter
undersample = RandomUnderSampler(sampling_strategy='majority', random_state = 42)
```

```python
Counter(data_cvd_death)
```

```python
# fit and transform
```

```python
x_under, y_under = undersample.fit_resample(x,data_cvd_death)
```

```python
Counter(y_under)
```

```python
x_under.to_csv("x_under_cvd_death.csv")
```
